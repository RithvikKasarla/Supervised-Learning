```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('DATA/sonar.all-data.csv')
```

```python
df.head()
```

```python
plt.figure(figsize = (6,4), dpi = 100)
sns.heatmap(df.drop('Label',axis = 1).corr())
```

```python
new_df = pd.get_dummies(df,drop_first=True)
```

```python
new_df
```

```python
X = new_df.drop('Label_R',axis = 1)
y = new_df['Label_R']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size= .1, random_state = 42)
```

```python
from sklearn.preprocessing import StandardScaler
scalar = StandardScaler()
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier()
```

```python
from sklearn.pipeline import Pipeline
operations = [('scalar', scalar),('knn',knn)]
pipe = Pipeline(operations)
```

```python
from sklearn.model_selection import GridSearchCV
k_vals = list(range(1,20))
param_grid = {'knn__n_neighbors': k_vals }
cv_classifier = GridSearchCV(pipe, param_grid, cv = 5, scoring = 'accuracy')
cv_classifier.fit(X_train, y_train)
cv_classifier.best_params_
```

```python
plt.plot(k_vals,cv_classifier.cv_results_['mean_test_score'],'o-')
```

```python
from sklearn.metrics import classification_report, confusion_matrix
print(confusion_matrix(cv_classifier.predict(X_test),y_test))
print(classification_report(cv_classifier.predict(X_test),y_test))
```
